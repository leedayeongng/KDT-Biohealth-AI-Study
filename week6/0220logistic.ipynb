```python
#x:공부시간, y:합격[1]/불합격[0]
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(p_x):
    return 1 / (1 + np.exp(-p_x))

x = np.array([2, 4, 6, 8, 10, 12, 14])
y = np.array([0, 0, 0, 1, 1, 1, 1])
```

```python
test_x = np.linspace(-10, 10, 400) # -10~ 10 랜덤값 400개
test_y = sigmoid(test_x)
plt.plot(test_x, test_y)
plt.show()
```

```python
a= 0
b = 0
lr = 0.05
epochs = 2001
n = len(x)
for i in range(epochs):
    y_hat = sigmoid(a*x+b) # 예측
    error = y - y_hat
    a_diff = -(1/n) * sum(x * error)
    b_diff = -(1/n) * sum(error)
    a = a - lr * a_diff
    b = b - lr * b_diff
    if i % 100 == 0:
        print(f'epochs={i}, a={a}, b={b}')
plt.scatter(x,y)
x_range = np.arange(0, 15, 0.1)
plt.plot(x_range, sigmoid(a * x_range + b))
plt.show()
```
