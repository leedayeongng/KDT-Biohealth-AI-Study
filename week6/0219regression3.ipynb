```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#공부시간 : X1, 과외:x2  성적:Y
x1 = np.array([2, 4, 6, 8])
x2 = np.array([0, 4, 2, 3])
y = np.array([81, 93, 91, 97])
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x1, x2, y)
```

```python
# 임의의 기울기 a, y절편 b
a1 = 0
a2 = 0
b = 0
n = len(x1)
# 학습률
lr = 0.001
#몇번 학습
epochs = 3001
for i in range(epochs):
    y_hat = a1 * x1 + a2 * x2 + b #예측
    error = y - y_hat #오차
    a1_diff = (2/(n)) * sum(-x1 * (error)) # 오차함수를 a로 편미분
    a2_diff = (2/(n)) * sum(-x2 * (error))
    b_diff = (2/n) * sum(-(error)) # 오차함수를 b로 편미분
    a1 = a1 - lr * a1_diff # 업데이트
    a2 = a2 - lr * a2_diff
    b = b - lr * b_diff #업데이트
    if i % 100 == 0:
        print(f'epochs: {i}, 기울기a1={a1}, 기울기a2={a2}, 절편={b}')
```

```python
# 3d 그래프
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x1, x2, y, color='red', label='actual data')
#예측평면
x1_grid, x2_grid = np.meshgrid(
    np.linspace(min(x1), max(x1), 10), np.linspace(min(x2), max(x2), 10)
)
y1_grid = a1 * x1_grid + a2 * x2_grid + b
ax.plot_surface(x1_grid, x2_grid, y1_grid, alpha=0.5, cmap='viridis')
ax.set_xlabel('study hour')
ax.set_ylabel('private hour')
ax.set_zlabel('score(y)')
ax.legend()
```
