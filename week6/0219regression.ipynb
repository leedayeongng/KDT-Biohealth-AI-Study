```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#공부시간 : X, 성적:Y
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])
plt.scatter(x,y)
```

```python
# 임의의 기울기 a, y절편 b
a = 0
b = 0
# 학습률
lr = 0.01
#몇번 학습
epochs = 2001
for i in range(epochs):
    y_hat = a * x + b #예측
    error = y - y_hat #오차
    a_diff = (2/len(x)) * sum(-x * (error)) # 오차함수를 a로 편미분
    b_diff = (2/len(x)) * sum(-(error)) # 오차함수를 b로 편미분
    a = a - lr * a_diff # 업데이트
    b = b - lr * b_diff #업데이트
    if i % 100 == 0:
        print(f'epochs: {i}, 기울기={a}, 절편={b}')
```

```python
y_pred = a * x + b
plt.scatter(x,y)
plt.plot(x,y_pred, 'r')
```
