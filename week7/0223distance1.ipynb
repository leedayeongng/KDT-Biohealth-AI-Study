```python
#L1 노름 : 백터의 모든 요소의 절댓값, 각자 형태로 이동(맨해튼거리)
#L2 노름 : 백터의 각 요소의 제곱창에 루트를 씌운, 횡단 직선거리(유클라드 거리)
import numpy as np
import matplotlib.pyplot as plt
a = np.array([1, 1])
b = np.array([5, 4])

manhattan = np.linalg.norm(a -b, ord=1)
euclidean = np.linalg.norm(a - b, ord=2)

manhattan, euclidean

fig, ax = plt.subplots(figsize=(6, 6))
ax.scatter(a[0], a[1], color='tab:blue', s=70, label='point A')
ax.scatter(b[0], b[1], color='tab:orange', s=70, label='point B')
#직선
ax.plot([a[0], b[0]], [a[1], b[1]] , 'g--', linewidth=2, label=f'euclidean={euclidean:.2f}')
#격자
mid = np.array([b[0],a[1]])
ax.plot([a[0], mid[0]], [a[1], mid[1]], color='red', linewidth=2)
ax.plot([mid[0], b[0]], [mid[1], b[1]] , 'red', linewidth=2, label=f'manhattan={manhattan:.2f}')

ax.set_title('L2 vs L1')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
plt.show()
```
