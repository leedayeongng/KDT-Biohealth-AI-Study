```python
import joblib
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report, accuracy_score

model = joblib.load('loan_model.pkl')
model_features = joblib.load('model_features.pkl')
test_data = pd.read_csv('../datasets/decision/test_loan_20.csv')
test_encoded = pd.get_dummies(test_data, drop_first=True)
#학습 컬럼 순서로
x = test_encoded.reindex(columns=model_features, fill_value=0)
print(x)
test_pred = model.predict(x)
test_data['loan_pred'] = np.where(test_pred == 1, 'Approved', 'Denied')
test_data.to_csv('result.csv', index=False)

y = test_encoded['Loan_Status_Y']
acc = accuracy_score(y, test_pred)
print('acc',acc)
```
