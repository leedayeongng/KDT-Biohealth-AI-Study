### LaTex 문서 조판 언어로, 논문이나 출판물 등의 특수 형식 문서를 작성하는데 쓰임
- 거듭 제곱
$$ y=x^3 $$

- 분수
$$ y=\frac{1}{2}x + 1$$



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def my_func(x):
    a=3
    return x**a # x의 a제곱
x= np.linspace(0,2)
y=my_func(x)
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
def my_func(x):
    return np.sqrt(x) + 1
x=np.linspace(0,4)
y=my_func(x)
plt.plot(x,y)
plt.grid()
plt.show()
```

$$ y=\sqrt x + 1 $$

$$ y=3x^2 + 2^x +1 $$

```python
def my_func(x):
    return 3*x**2 + 2*x + 1
x=np.linspace(-2,2)
y=my_func(x)
plt.plot(x,y)
plt.grid()
plt.show()
```

#3차원 다향식을 그래프로 그리시오

```python
def my_func(x):
    return 4*x**3 + 2*x**2 + x + 3
x=np.linspace(-2,2)
y=my_func(x)
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
###다형식 차수별 그래프
np.random.seed(42)
x = np.linspace(0, 2*np.pi, 20)  # 0-2파이 사이를 20개 점으로 나눔
y = np.sin(x) + np.random.normal(0,0.2, len(x))
y_plot = np.linspace(0,2*np.pi, 100)
plt.figure(figsize=(16,10))
degrees = [1, 2, 3, 9]
color = ['red', 'green', 'orange', 'purple']
for i, degrees in enumerate(degrees):
    # 해당 차수에 맞는 다향식 계수 도축
    cofficients = np.polyfit(x,y,degrees)
    #도출된 파라미터로 함수식 조합(f(x))
    p = np.poly1d(cofficients)
    plt.subplot(2,2,i+1)
    plt.scatter(x,y, color='blue')
    plt.plot(y_plot, p(y_plot), color=color[i])
plt.tight_layout()
plt.show()
```
