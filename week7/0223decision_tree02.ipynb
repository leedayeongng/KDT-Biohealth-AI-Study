```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree #학습된나무를그림으로보여줌
from IPython.display import display
from sklearn.datasets import load_iris #머신러닝 교과서 데이터셋
from sklearn.model_selection import train_test_split

iris=load_iris()
x_train, x_test, y_train, y_test = train_test_split(iris['data'], iris['target'], test_size=0.2, random_state=2)
#값이 제한없이
deep_model = DecisionTreeClassifier(random_state=3)
deep_model.fit(x_train, y_train)
print('값이 제한 없이')
print(f'훈련:{deep_model.score(x_train, y_train)}')
print(f'테스트:{deep_model.score(x_test, y_test)}')

plt.figure(figsize=(12,8))
plot_tree(deep_model, filled=True, rounded=True)
plt.show()
```

```python
#max_depth 설정 (pruning 가지치기)
prune_model = DecisionTreeClassifier(max_depth=3, random_state=2)
prune_model.fit(x_train, y_train)
print('가지치기 후 max_depth=3')
print(f'훈련:{prune_model.score(x_train, y_train)}')
print(f'테스트:{prune_model.score(x_test, y_test)}')
plt.figure(figsize=(12,6))
plot_tree(prune_model, feature_names=iris['feature_names'], class_names=iris['target_names'], filled=True, rounded=True)
plt.show()
```
